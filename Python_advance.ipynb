File handling

The open() function is constructure of File object

The with statement relies on two magic methods: __enter__() and __exit__().

__enter__(): Executed when the with block starts. It opens the file and returns the file object.

__exit__(type, value, traceback): Executed when the block ends (or if an error occurs). It automatically closes the file.

```python
f = open("data.txt", "w")
try:
    f.write("Hello")
finally:
    f.close()

# safe way
with open("data.txt", "w") as f:
    f.write("Hello")
```

```python
with open("data.txt", "r") as f:
    content=f.read()
    print(content)
```

Python uses internal buffer when writing, first writes into RAM and then writes the entire chunk into Disk when file closes.
