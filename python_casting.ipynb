#### **IF statements**

if statements allow your program to make decisions based on conditions.
Python evaluates conditions as True or False and executes the matching block.

```python
score = 72

if score >= 90:
    grade = "A"
elif score >= 70:
    grade = "B"
else:
    grade = "C"

print(grade)
```

#### **for Loop – Iteration Over Sequences**

A for loop is used to iterate over iterable objects like lists, strings, tuples, sets, dictionaries, and ranges.

```python
numbers = [2, 4, 6, 8]

for num in numbers:
    print(num * num)
```
